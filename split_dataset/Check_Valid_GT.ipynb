```python
import imageio as io
import matplotlib.pyplot as plt

number = 2746 

str_img = './test/data/BG_SYSU_' + str(number) + '.jpg'
img = io.imread(str_img)
meta = open('./test/gt.txt', 'r')

for i, line in enumerate(meta):
    line = line.split(' ')
    if int(line[0]) == number:
        data = line


img_height = img.shape[0]
img_width = img.shape[1]

# This code works for single-vehicle images, 
vehicle_x = int(data[14])
vehicle_y = int(data[15])
width = int(data[16])
height = int(data[17])

print(data)       
print("img shape: ", img.shape)
print(vehicle_x, vehicle_y, width, height)

car_img = img[vehicle_y:(vehicle_y + height),
              vehicle_x:(vehicle_x + width)]
               
    
print('width: ' + str(width))    
    
fig = plt.figure(figsize=(20, 20))
plt.subplot(1, 2, 1)
plt.imshow(img)

plt.subplot(1, 2, 2)
plt.imshow(car_img)
```
